```python
# crear etiquetas y agregarlas al DOM

from bs4 import BeautifulSoup

if __name__=='__main__': 

    with open ('econpy.html', 'r') as f:  
        content = f.read()

        soup = BeautifulSoup(content, 'html.parser') 

        a= soup.new_tag('a', href = 'www.codigofacilito.com') #crear etiqueta
        a.string = 'Link a la plataforma' #agregar texto a la etiqueta
        div = soup.new_tag('div', id = 'item-01', title= 'item-data') # crear etiqueta
        div['nuevo-atributo'] = 'nuevo-atributo'  # agregar atributo
        div.append('\n')
        div.append(a) # agregar etiqueta a la etiqueta
        div.append('\n')
        
        soup.body.insert(1, div) # insertar etiqueta en el DOM

        print(soup.prettify()) # imprimir el DOM
# los nodos se agregaron de forma exitosa
```
